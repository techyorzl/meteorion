```python
import xarray as xr
import pandas as pd

paramIds = [165, 166, 168, 167, 151, 34, 134]
dataframes = {}

for paramId in paramIds:
    try:
        ds = xr.open_dataset('data.grib', engine='cfgrib', filter_by_keys={'paramId': paramId})
        df = ds.to_dataframe().reset_index()
        df['paramId'] = paramId
        dataframes[paramId] = df
        print(f"DataFrame for paramId {paramId}:")
        print(df.head(), "\n")

    except Exception as e:
        print(f"Error processing paramId {paramId}: {e}")
combined_df = pd.concat(dataframes.values(), ignore_index=True)

print("Combined DataFrame:")
print(combined_df.head())
```
